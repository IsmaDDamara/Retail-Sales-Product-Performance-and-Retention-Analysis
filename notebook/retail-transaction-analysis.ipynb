## **DATA UNDERSTANDING**

### **Deskripsi Variabel dalam Dataset**


1. **`id`**: ID unik untuk setiap entri transaksi.  

2. **`customer_id`**: ID unik pelanggan.  

3. **`order_date`**: Tanggal transaksi dilakukan.  

4. **`sku_id`**: ID unik produk berdasarkan SKU.  

5. **`price`**: Harga satuan produk.  

6. **`qty_ordered`**: Jumlah unit produk yang dipesan.  

7. **`before_discount`**: Total harga sebelum diskon diterapkan.  

8. **`discount_amount`**: Jumlah potongan harga pada transaksi.  

9. **`after_discount`**: Total harga setelah diskon diterapkan.  

10. **`is_gross`**: Indikator apakah harga merupakan nilai bruto.

11. **`is_valid`**: Indikator apakah transaksi valid atau tidak.  

12. **`is_net`**: Indikator apakah harga merupakan nilai netto.  

13. **`payment_id`**: ID unik metode pembayaran.

14. **`payment_method`**: Jenis metode pembayaran pada transaksi.

15. **`sku_name`**: Nama produk sesuai SKU.

16. **`base_price`**: Harga dasar produk sebelum ada diskon.  

17. **`cogs`**: Biaya produksi barang yang terjual (COGS).  

18. **`category`**: Kategori produk berdasarkan jenisnya.  

19. **`registered_date`**: Tanggal pelanggan mendaftar dalam sistem.

## **DATA CLEANSING**

### **Data Transformation**



Mengubah tipe data menjadi yang seharusnya untuk memudahkan proses analisis

```python
df_clean = df.copy()



# Mengubah tipe data kolom 'date' dan 'registered_date' menjadi datetime

df_clean['order_date'] = pd.to_datetime(df_clean['order_date'])

df_clean['registered_date'] = pd.to_datetime(df_clean['registered_date'])
```

```python
df_clean.info()
```

### **Handling Negative Value**

```python
# Mengecek apakah terdapat nilai minus pada dataset int

df_clean.select_dtypes(include='int64').lt(0).sum()
```

```python
# Melihat data dengan nilai discount_amount negatif

df_clean[df_clean['discount_amount'] < 0]
```

```python
baris_data = 5883



# Mengubah nilai negatif 'discount_amount' pada baris ke-5883 menjadi positif

df_clean.loc[baris_data, 'discount_amount'] = abs(df_clean.loc[baris_data, 'discount_amount'])



# Menghitung ulang nilai 'after_discount' berdasarkan harga setelah diskon

df_clean.loc[baris_data, 'after_discount'] = df_clean.loc[baris_data, 'price'] - df_clean.loc[baris_data, 'discount_amount']
```

```python
df_clean
```

```python
df_clean.describe()
```

### **Menambahkan Kolom Baru**



Menambahkan kolom profit sebagai informasi keuntungan yang diterima.

```python
# Hitung profit

df_clean['profit'] = (df_clean['base_price'] - df_clean['cogs']) * df_clean['qty_ordered']
```

## **EXPLANATORY DATA ANALYSIS**

### **Identifikasi Tren Penjualan**

```python
# Copy dataset

data = df_clean.copy()



# Grup df_clean berdasarkan bulan dan hitung jumlah transaksi per bulan

data['month'] = data['order_date'].dt.to_period('M')

monthly_sales = data.groupby('month').size()



# Buat visualisasi

plt.figure(figsize=(10, 6))

plt.plot(monthly_sales.index.astype(str), monthly_sales.values, marker='o', color='pink', label='Transaksi')



# Tambahkan judul dan label

plt.title('Tren Jumlah Penjualan Bulanan', fontsize=16)

plt.xlabel('Bulan', fontsize=12)

plt.ylabel('Transaksi', fontsize=12)

plt.xticks(rotation=45)

plt.legend()

plt.grid(axis='y', linestyle='--', alpha=0.7)



# Tampilkan grafik

plt.tight_layout()

plt.show()
```

**Insight:**



Jumlah transaksi bulanan menunjukkan pola yang menarik selama periode analisis. Dari Januari hingga Mei 2021, jumlah transaksi relatif stabil di bawah 100 transaksi per bulan. Namun, mulai Juni 2021, terjadi lonjakan tajam yang mencapai puncaknya pada Juli 2021 dengan lebih dari 300 transaksi. Tren ini kemudian stabil di kisaran angka yang sama hingga pertengahan 2022. Sayangnya, mulai Juli 2022, jumlah transaksi mengalami penurunan signifikan dan terus menurun hingga akhir tahun 2022. Pola ini menunjukkan adanya faktor tertentu di pertengahan 2021 yang mendorong kenaikan transaksi, diikuti oleh faktor lain di tahun 2022 yang menyebabkan penurunan drastis.

**Insight:**



Pendapatan bulanan mencerminkan tren transaksi, dengan peningkatan drastis yang dimulai pada Juni 2021 dan mencapai puncaknya di pertengahan tahun tersebut. Setelah itu, pendapatan tetap stabil di kisaran yang cukup tinggi hingga pertengahan 2022. Hal ini menunjukkan bahwa pola pembelian pelanggan sangat memengaruhi pendapatan, dan kemungkinan adanya faktor eksternal seperti penurunan daya beli atau perubahan strategi bisnis yang berkontribusi pada penurunan tersebut.

## **USER RETENTION RATE**



User Retention Rate adalah metrik yang digunakan untuk mengukur persentase pelanggan yang kembali melakukan pembelian pada periode berikutnya, seperti bulan atau tahun, guna mengevaluasi tingkat loyalitas pelanggan dan efektivitas strategi retensi.

### **Retention Rate Data Tahun 2021**

#### **Agregat data transaksi ke bentuk summary total transaksi/order setiap pengguna setiap bulan**



Menghitung total pesanan yang dilakukan oleh setiap pelanggan setiap bulan, untuk melihat seberapa sering mereka melakukan transaksi di setiap bulan.

```python
# Salin data dan tambahkan kolom 'year_month' dan 'year'

df_retention_2021 = df_clean.copy()



# Tambahkan kolom 'year_month' berdasarkan 'order_date'

df_retention_2021['year_month'] = df_retention_2021['order_date'].dt.to_period('M')



# Tambahkan kolom 'year' untuk mempermudah penyaringan tahun

df_retention_2021['year'] = df_retention_2021['order_date'].dt.year



# Hitung jumlah pesanan unik per pengguna per bulan

df_retention_2021 = df_retention_2021.groupby(['customer_id', 'year_month'], as_index=False).agg(order_cnt=('id', 'nunique'))



# Pisahkan cohort untuk tahun 2021

df_2021 = df_retention_2021[df_retention_2021['year_month'].dt.year == 2021]



# Tampilkan data tahun 2021

df_2021.head()
```

#### **Membuat kolom sebagai cohort dari pengguna**



Menambahkan kolom cohort yang berisi bulan pertama (min) setiap pelanggan (customer_id) berdasarkan year_month. Untuk mengetahui bulan pertama saat melakukan transaksi pada masing-masing customer.

```python
df_2021['cohort'] = df_2021.groupby('customer_id')['year_month'].transform('min')



df_2021.head()
```

#### **Menghitung Jarak Bulan Transaksi dari Bulan Pertama dan Menambahkan 1 untuk Menandakan Bulan Pertama.**



Menghitung selisih bulan antara bulan transaksi dengan bulan pertama kali transaksi untuk setiap pelanggan, kemudian menambahkan 1 pada hasil perhitungan agar bulan pertama transaksi dihitung sebagai bulan 1, bulan kedua sebagai bulan 2, dan seterusnya. Ini bertujuan untuk memberikan urutan bulan yang lebih jelas berdasarkan periode transaksi pertama setiap pelanggan.

```python
from operator import attrgetter # Mengimpor fungsi untuk mengambil atribut objek.



# Menghitung selisih bulan antara transaksi dan bulan pertama pembelian

# (cohort), lalu ditambah 1 agar periode dimulai dari 1.

df_2021['period_num'] = (df_2021['year_month'] - df_2021['cohort']).apply(attrgetter('n')) + 1



# Menampilkan DataFrame dengan kolom 'period_num' yang baru ditambahkan.

df_2021.head()
```

####**Membuat Tabel Pivot untuk Pengguna Unik Berdasarkan Cohort dan Jarak Bulan**



Tabel pivot ini menggunakan cohort sebagai indeks, jarak bulan sebagai kolom, dan nilai yang dihitung adalah jumlah pengguna unik (berdasarkan ID pengguna) untuk setiap kombinasi cohort dan jarak bulan.

```python
# membuat tabel pivot

df_cohort_pivot_2021 = pd.pivot_table(df_2021, index='cohort', columns='period_num', values='customer_id', aggfunc=pd.Series.nunique)

df_cohort_pivot_2021
```

####**Menghitung Retention Rate Berdasarkan Cohort Pengguna**



Menghitung jumlah pengguna yang pertama kali bertransaksi di setiap bulan (cohort), lalu membagi nilai pada tabel pivot dengan jumlah pengguna di masing-masing cohort untuk mendapatkan retention rate, yang menunjukkan persentase pengguna yang kembali bertransaksi di bulan-bulan berikutnya.

```python
# Jumlah pengguna unik di setiap cohort (periode pertama).

cohort_size_2021 = df_cohort_pivot_2021.iloc[:, 0]



cohort_size_2021
```

```python
# Menghitung retention rate per cohort.

df_retention_cohort_2021 = df_cohort_pivot_2021.divide(cohort_size_2021, axis=0)

df_retention_cohort_2021
```

#### **Menampilkan Heatmap Retention Rate**



Heatmap digunakan untuk memvisualisasikan retention rate dari tabel pivot, dengan warna yang menggambarkan tingkat retensi pengguna pada setiap kombinasi cohort dan jarak bulan.

### **Retention Rate Data Tahun 2022**

#### **Agregat data transaksi ke bentuk summary total transaksi/order setiap pengguna setiap bulan**



Menghitung total pesanan yang dilakukan oleh setiap pelanggan setiap bulan, untuk melihat seberapa sering mereka melakukan transaksi di setiap bulan.

```python
# Salin data dan tambahkan kolom 'year_month' dan 'year'

df_retention_2022 = df_clean.copy()



# Tambahkan kolom 'year_month' berdasarkan 'order_date'

df_retention_2022['year_month'] = df_retention_2022['order_date'].dt.to_period('M')



# Tambahkan kolom 'year' untuk mempermudah penyaringan tahun

df_retention_2022['year'] = df_retention_2022['order_date'].dt.year



# Hitung jumlah pesanan unik per pengguna per bulan

df_retention_2022 = df_retention_2022.groupby(['customer_id', 'year_month'], as_index=False).agg(order_cnt=('id', 'nunique'))



# Pisahkan cohort untuk tahun 2021

df_2022 = df_retention_2022[df_retention_2022['year_month'].dt.year == 2022]



# Tampilkan data tahun 2021

df_2022.head()
```

#### **Membuat kolom sebagai cohort dari pengguna**



Menambahkan kolom cohort yang berisi bulan pertama (min) setiap pelanggan (customer_id) berdasarkan year_month. Untuk mengetahui bulan pertama saat melakukan transaksi pada masing-masing customer.

```python
df_2022['cohort'] = df_2022.groupby('customer_id')['year_month'].transform('min')



df_2022.head()
```

#### **Menghitung Jarak Bulan Transaksi dari Bulan Pertama dan Menambahkan 1 untuk Menandakan Bulan Pertama.**



Menghitung selisih bulan antara bulan transaksi dengan bulan pertama kali transaksi untuk setiap pelanggan, kemudian menambahkan 1 pada hasil perhitungan agar bulan pertama transaksi dihitung sebagai bulan 1, bulan kedua sebagai bulan 2, dan seterusnya. Ini bertujuan untuk memberikan urutan bulan yang lebih jelas berdasarkan periode transaksi pertama setiap pelanggan.

```python
from operator import attrgetter # Mengimpor fungsi untuk mengambil atribut objek.



# Menghitung selisih bulan antara transaksi dan bulan pertama pembelian

# (cohort), lalu ditambah 1 agar periode dimulai dari 1.

df_2022['period_num'] = (df_2022['year_month'] - df_2022['cohort']).apply(attrgetter('n')) + 1



# Menampilkan DataFrame dengan kolom 'period_num' yang baru ditambahkan.

df_2022.head()
```

####**Membuat Tabel Pivot untuk Pengguna Unik Berdasarkan Cohort dan Jarak Bulan**



Tabel pivot ini menggunakan cohort sebagai indeks, jarak bulan sebagai kolom, dan nilai yang dihitung adalah jumlah pengguna unik (berdasarkan ID pengguna) untuk setiap kombinasi cohort dan jarak bulan.

```python
# membuat tabel pivot

df_cohort_pivot_2022 = pd.pivot_table(df_2022, index='cohort', columns='period_num', values='customer_id', aggfunc=pd.Series.nunique)

df_cohort_pivot_2022
```

####**Menghitung Retention Rate Berdasarkan Cohort Pengguna**



Menghitung jumlah pengguna yang pertama kali bertransaksi di setiap bulan (cohort), lalu membagi nilai pada tabel pivot dengan jumlah pengguna di masing-masing cohort untuk mendapatkan retention rate, yang menunjukkan persentase pengguna yang kembali bertransaksi di bulan-bulan berikutnya.

```python
# Jumlah pengguna unik di setiap cohort (periode pertama).

cohort_size_2022 = df_cohort_pivot_2022.iloc[:, 0]



cohort_size_2022
```

```python
# Menghitung retention rate per cohort.

df_retention_cohort_2022 = df_cohort_pivot_2022.divide(cohort_size_2022, axis=0)

df_retention_cohort_2022
```

#### **Menampilkan Heatmap Retention Rate**



Heatmap digunakan untuk memvisualisasikan retention rate dari tabel pivot, dengan warna yang menggambarkan tingkat retensi pengguna pada setiap kombinasi cohort dan jarak bulan.

```python
import matplotlib.pyplot as plt

import matplotlib.colors as mcolors

import seaborn as sns

import pandas as pd



# Warna hijau muda untuk colormap

# Warna biru muda untuk colormap

light_blue_cmap = mcolors.ListedColormap(['#e6faff', '#ccf2ff', '#99e6ff', '#66d9ff', '#33ccff', '#00bfff'])



# Set custom styles for aesthetics

sns.set(style="white", font_scale=1.1)



with sns.axes_style('white'):

    fig, ax = plt.subplots(1, 2, figsize=(14, 8), sharey=True, gridspec_kw={'width_ratios': [1, 11]})



    # User retention cohort with green colormap

    sns.heatmap(

        df_retention_cohort_2022, annot=True, fmt='.0%', cmap=light_blue_cmap,

        ax=ax[1], cbar=False, linewidths=0.1, linecolor='grey',

        annot_kws={"size": 10, "weight": "bold", "color": "black"}

    )

    ax[1].set_title('User Retention Cohort 2022', fontsize=16, fontweight='bold')

    ax[1].set(xlabel='Month Number', ylabel='')



    # Cohort size with a lighter blue background

    df_cohort_size = pd.DataFrame(cohort_size_2022)

    light_blue_cmap = mcolors.ListedColormap(['#ff80ab'])  # Light blue color

    sns.heatmap(

        df_cohort_size, annot=True, cbar=False, fmt='g', cmap=light_blue_cmap, ax=ax[0],

        annot_kws={"size": 10, "weight": "bold", "color": "black"}

    )

    ax[0].tick_params(bottom=False)

    ax[0].set(xlabel='Cohort Size', ylabel='First Order Month', xticklabels=[])

    ax[0].set_title('Cohort Size', fontsize=12, fontweight='bold')



    # Adjust layout for better spacing

    fig.tight_layout(pad=2.0)



plt.show()
```

**Insight:**



User Retention Cohort 2022 memperlihatkan pola penurunan retensi yang lebih drastis dibandingkan tahun sebelumnya. Meskipun retensi bulan pertama tetap 100%, angka ini turun jauh lebih tajam setelah bulan kedua. Misalnya, cohort Januari 2022 memiliki retensi sebesar 13% pada bulan ke-2, tetapi hanya 3% pada bulan ke-6. Cohort Mei 2022 bahkan menunjukkan penurunan retensi menjadi 1% atau kurang setelah bulan ke-3. Perbedaan mencolok dengan tahun 2021 terlihat dari rendahnya retensi rata-rata di tahun 2022, dengan sebagian besar cohort memiliki retensi di bawah 5% setelah bulan ke-3. Penurunan ini kemungkinan disebabkan oleh persaingan pasar yang lebih ketat, perubahan kebutuhan pelanggan, atau kualitas layanan yang tidak memenuhi harapan pengguna, sehingga mereka cenderung beralih ke platform lain.
